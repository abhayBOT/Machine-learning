```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/diabetes.csv', header=0)
```

```python
#to display top 5 
df.head()
```

```python
#drop null values
df = df.dropna()
```

```python
#display all feilds
print(list(df.columns))
```

```python
df.head()
```

```python
#To know column name
df.columns[3]
```

**SPLITING DATA**



```python
data = pd.get_dummies(df,columns= ['Pregnancies',	'DiabetesPedigreeFunction',	'Age',	'Outcome'])

X = data.iloc[:,1:]
```

```python
data.head()
```

```python
data.columns
```

```python

data
```

```python
#to create an array for predicted value column
Y = data.iloc[:,0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
```

**Building classifier**

```python
classifier = LogisticRegression(solver='lbfgs',random_state=0)
```

```python
classifier.fit(X_train, Y_train)
```

**TESTING**

```python
predicted_y = classifier.predict(X_test)
```

```python
predicted_y
```

```python
for x in range(len(predicted_y)):
   if (predicted_y[x] == 1):
      print(x, end="\t")
```

**ACCURACY**

```python
print('Accuracy: {:.2f}'.format(classifier.score(X_test, Y_test)))
```
